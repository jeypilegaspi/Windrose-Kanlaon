```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from datetime import datetime, timedelta
from matplotlib.ticker import FuncFormatter

# Load data from JSON
file_path = 'windrose_data_hres.json'

# Base class to process wind data
class WindDataProcessor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.wind_data = self.load_data()

    def load_data(self):
        with open(self.file_path, 'r') as file:
            return json.load(file)

    def filter_by_date_range(self, date_range):
        start_date, end_date = date_range
        filtered_data = []
        for entry in self.wind_data['windrose']:
            entry_date = datetime.strptime(entry['date'], "%B %d, %Y")
            if start_date <= entry_date <= end_date:
                filtered_data.append(entry)
        return filtered_data

    def get_data_by_altitude(self, altitude, date_range):
        data_by_direction = {d: [] for d in self.get_directions()}
        date_time_info = []  # List to store date and time information
        filtered_data = self.filter_by_date_range(date_range)
        for entry in filtered_data:
            entry_date = datetime.strptime(entry['date'], "%B %d, %Y")
            for measurement in entry['data']:
                if measurement['z_hPa'] == altitude:
                    direction = measurement['direction']
                    speed = measurement['speed_kph']
                    data_by_direction[direction].append(speed)
                    date_time_info.append((entry_date, entry['time']))  # Correctly store date and time
        return data_by_direction, date_time_info 

    @staticmethod
    def get_directions():
        return ['E', 'ENE', 'NE', 'NNE', 'N', 'NNW', 'NW', 'WNW', 'W', 'WSW', 'SW', 'SSW', 'S', 'SSE', 'SE', 'ESE']


# Subclass for plotting wind rose
class WindRosePlotter(WindDataProcessor):
    def __init__(self, file_path):
        super().__init__(file_path)

    def plot_windrose(self, altitude, date_range):
        data_by_direction, date_time_info = self.get_data_by_altitude(altitude, date_range)
        
        # Print all the date and time used
        print("Date and Time of Data Used:")
        for entry in date_time_info:
            print(f"Date: {entry[0].strftime('%Y-%m-%d')}, Time: {entry[1]}")

        # Prepare polar plot
        fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(12, 10))
        
        directions = self.get_directions()
        dir_dict = {d: i * 2 * np.pi / 16 for i, d in enumerate(directions)}
        
        # Define color map for different speed layers
        cmap = cm.get_cmap('Blues')
        
        # Initialize the array for magnitudes
        magnitude_layers = []
        
        # Prepare the data for stacking
        for direction in directions:
            speeds = data_by_direction[direction]
            if speeds:  # Only process if there are speeds available
                magnitude_layers.append((dir_dict[direction], speeds))
        
        # Sort layers based on speed for better stacking visualization
        magnitude_layers.sort(key=lambda x: np.mean(x[1]), reverse=True)

        # Calculate maximum number of speed layers for proper plotting
        max_layers = max(len(speeds) for _, speeds in magnitude_layers) if magnitude_layers else 0

        # Plot the stacked layers
        for dir_index, speeds in magnitude_layers:
            for layer in range(max_layers):
                if layer < len(speeds):  # Check if the layer exists
                    # Use layer index for different shades or colors
                    color = cmap(min(layer / 5, 1))  # Adjust as necessary for shades
                    bottom = np.sum([s[layer] for s in magnitude_layers[:magnitude_layers.index((dir_index, speeds))]], axis=0) if layer > 0 else 0
                    ax.bar(dir_index, speeds[layer], width=2 * np.pi / 16, 
                           bottom=bottom, color=color, edgecolor='k', alpha=0.8)

        # Customize radial labels and ticks
        ax.set_xticks(np.linspace(0.0, 2 * np.pi, len(directions), endpoint=False))
        ax.set_xticklabels(directions)
        ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{x:.1f} km/h'))
        
        # Add title and show plot
        ax.set_title(f"Windrose at {altitude} hPa", va='bottom')
        plt.show()

# Example Usage
file_path = 'windrose_data_hres.json'
plotter = WindRosePlotter(file_path)

# Specify a date range (7-day period) and altitude (e.g., 850 hPa)
start_date = datetime(2024, 6, 3)
end_date = start_date + timedelta(days=6)  # 7-day range
plotter.plot_windrose(850, (start_date, end_date))
plotter.plot_windrose(700, (start_date, end_date))
plotter.plot_windrose(500, (start_date, end_date))
plotter.plot_windrose(300, (start_date, end_date))
plotter.plot_windrose(200, (start_date, end_date))
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from matplotlib import colormaps
from datetime import datetime, timedelta

# Load data from the JSON file
file_path = 'windrose_data_hres.json'

with open(file_path, 'r') as file:
    data = json.load(file)

def plot_windrose_by_altitude(date_range=None):
    # Define directions and altitude labels
    directions = ['E', 'ENE', 'NE', 'NNE', 'N', 'NNW', 'NW', 'WNW', 'W', 'WSW', 'SW', 'SSW', 'S', 'SSE', 'SE', 'ESE']
    dir_dict = {d: i * 2 * np.pi / 16 for i, d in enumerate(directions)}
    altitude_levels = [850, 700, 500, 300, 200]  # Altitude levels in hPa
    altitude_labels = ["~1.5km", "~3km", "~5-6km", "~9km", "~12km"]
    
    # Prepare data for each altitude level over the 7-day range
    wind_data_by_altitude = {alt: {d: [] for d in directions} for alt in altitude_levels}
    
    # Filter data by the date range
    for entry in data['windrose']:
        entry_date = datetime.strptime(entry['date'], "%B %d, %Y")
        if date_range and (entry_date < date_range[0] or entry_date > date_range[1]):
            continue
        
        measurements = entry['data']
        
        # Collect wind speeds for each altitude level
        for measurement in measurements:
            direction = measurement['direction']
            speed = measurement['speed_kph']
            z_hPa = measurement['z_hPa']
            if z_hPa in wind_data_by_altitude:
                wind_data_by_altitude[z_hPa][direction].append(speed)
    
    # Plot 5 diagrams, one for each altitude level
    for alt_idx, alt in enumerate(altitude_levels):
        fig, ax = plt.subplots(figsize=(12, 10), subplot_kw={'polar': True})
        
        # Define color map and shades
        cmap = colormaps['Blues']  # Change 'Blues' to other colormap if you like
        shades = [cmap(i) for i in np.arange(0.1, 5, 0.1)]  # Different shades for stacking
        
        # Collect wind speeds and stack them with shades for each layer
        for direction, speeds in wind_data_by_altitude[alt].items():
            if len(speeds) > 0:
                speeds.sort()  # Sort the speeds to stack them correctly (light to dark)
                bottom = 0
                for i, speed in enumerate(speeds):
                    ax.bar(dir_dict[direction], speed, width=2 * np.pi / 16, 
                           bottom=bottom, color=shades[min(i, len(shades)-1)], edgecolor='k', alpha=0.8)
                    bottom += speed  # Stack the bars
        
        # Customize the radial and angular ticks
        def radius_formatter(value, pos):
            return f'{value:.1f} km/h'
        ax.yaxis.set_major_formatter(FuncFormatter(radius_formatter))
        ax.set_xticks(np.linspace(0.0, 2 * np.pi, len(directions), endpoint=False))
        ax.set_xticklabels(directions)
        
        ax.set_title(f"Windrose for {altitude_labels[alt_idx]} (~{alt} hPa)", va='bottom')
        
        plt.subplots_adjust(left=0.1, right=0.85, top=0.85, bottom=0.1)
        plt.show()

# Example usage:
# Specify a date range of 7 days
start_date = datetime(2024, 6, 3)
end_date = start_date + timedelta(days=6)  # 7-day range
plot_windrose_by_altitude((start_date, end_date))
```
